```python
tipo_datos_sqlalchemy = {
    'int64': types.BIGINT(),
    'float64': types.FLOAT(),
}

# Definir el tipo de datos predeterminado para columnas no clasificadas
default_data_type = types.NVARCHAR(255)

# Crear un diccionario de tipos de datos para usar con to_sql
tipos_de_datos_sqlalchemy = {}

for columna, tipo in df_1.dtypes.items():
    tipo_sqlalchemy = tipo_datos_sqlalchemy.get(str(tipo), default_data_type)
    tipos_de_datos_sqlalchemy[columna] = tipo_sqlalchemy

# Utilizar to_sql para importar el DataFrame con los tipos de datos de SQLAlchemy especificados
df_1.to_sql('retest', engine, if_exists='replace', index=False, dtype=tipos_de_datos_sqlalchemy)

```
